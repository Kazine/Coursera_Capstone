```python
# importing packages
import numpy as np 
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python

# assiging link
link='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# getting the contents of the website
html_content=urllib.request.urlopen(link).read()
soup=BeautifulSoup(html_content,"lxml")
print(soup.prettify())
```

```python
#getting only table data
tb=soup.find("table", attrs={"class":'wikitable sortable'})

# finding each row (tr) of the table 
table_rows = tb.find_all("tr")
table_rows
```

```python
# Creating Lists
headings = []
table_data = [] 

#extracting Table Headings
for th in tb.find_all("th"):
    headings.append(th.text.replace('/n','').strip(),)

# extracting each row from the data in the table 
for tr in table_rows:
    td=tr.find_all('td')
    row=[i.text.strip() for i in td]
    #print(row)
    table_data.append(row)
    


# Creating Data fram
df=pd.DataFrame(table_data)

# Labeling Columns
df.columns=headings
```

```python
# identifing unique values in Borough Column
df['Borough'].unique()
```

```python
#Dropping Cells where Borough is not assigned
# Creating copy so i dont overwrie
df1=df
#replacing Not assigned to no value ( No Neighborood that has Not Assigned has a Burough with a name)

df1.replace({'Not assigned': None}, inplace=True)
# dropping Boroughs with no value if there is another NA
df1.dropna(thresh=2,inplace=True)

# Checking if any NA left in Neighborhoood
NA=df1['Neighborhood'].isna()
NA.unique()
```

## Defining command to get venues for each postal code

```python
def getNearbyVenues(names,neighborhoods, latitudes, longitudes, radius=500):  
    # Output 
    venues_list=[]
    for name,neighborhood, lat, lng in zip(names,neighborhoods,latitudes, longitudes):
        print(name,neighborhood)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name,
            neighborhood,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code',
                             'Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues=getNearbyVenues(names=toronto_data['Postal Code'],neighborhoods=toronto_data['Neighborhood'],latitudes=toronto_data['Latitude'],longitudes=toronto_data['Longitude'])
```

## Creating Dummy Table to get top venues for each section

```python


# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues1[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Postal Code'] = toronto_venues1['Postal Code'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
toronto_grouped = toronto_onehot.groupby('Postal Code').mean().reset_index()
toronto_grouped
```

```python
# Examining the frequency in each Postal Code

num_top_venues = 5

for hood in toronto_grouped['Postal Code']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Postal Code'] == hood].T.reset_index()
    temp.columns = ['Venue Category','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# making a command to locate top venues for each neighborhood

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 8 # number of venues

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Postal Code']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

```

```python
# create a new dataframe with top Venues 
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Postal Code'] = toronto_grouped['Postal Code']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted

# set number of clusters
kclusters = 6

toronto_grouped_clustering = toronto_grouped.drop('Postal Code', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Postal Code'), on='Postal Code')



```
